### Scrap Star wars character from GraphQL API

```python
import requests
import json
import pandas as pd

url = "https://swapi-graphql.netlify.app/.netlify/functions/index"

payload = "{\"query\":\"{\\r\\n  allPeople {\\r\\n    people {\\r\\n      name,\\r\\n      birthYear,\\r\\n      gender,\\r\\n      eyeColor,\\r\\n      hairColor,\\r\\n      height,\\r\\n      mass\\r\\n    }\\r\\n  }\\r\\n}\",\"variables\":{}}"
headers = {
  'Content-Type': 'application/json'
}

response = requests.request("POST", url, headers=headers, data=payload)

print(response.text)
```

```python
response.json()
```

```python
df_star_wars = pd.DataFrame(response.json()['data']['allPeople']['people'])
df_star_wars.columns = ['name','birth year', 'gender', 'eye color', 'hair color', 'height', 'mass']
df_star_wars.index += 1
df_star_wars
```

```python
df_star_wars.to_csv("Star Wars Character.csv", index = True)
```
